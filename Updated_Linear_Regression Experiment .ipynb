# Introduction:

## We've all seen rivers before, but where do they come from? Where do they go? How affected are they by changes in precipitation and evaporation rates? The baseflow of a river is a measure of the amount of the volume of water that a river is moving at any given time. Using data collected by the fine folks at the WATS department, we want to explore which factors (if any) affect baseflow significantly. We also want to know if these factors affect different segments of a river differently. Finally, we want to know if we can predict the amount of baseflow a river segment will produce given another attribute or combination of attributes.



# Dataset

# Each row of the dataset corresponds to one measure of a river segment during one month. The dataset contains the following columns:

### 1) Date: number of days since 01/01/0000
### 2) Segment ID: a categorical identifier of the specific segment of river
### 3) X/Y: spatial location of the station that observes the baseflow
### 4) Evapotranspiration: a measure of the amount of water that leaves the ground and plants through evaporation and transpiration in the area surrounding the river segment
### 5) Precipitation: a measure of the amount of water that has fallen from the atmosphere to the ground in the area surrounding the river segment
### 6) Irrigation Pumping: the amound of groundwater pumped out for irrigation in the area surrounding the river segment
### 7) Observed: the observed baseflow. The target variable!

### We have carefully analysed whether there is any null value present or not. All the data having null values is removed  and surprisingly we found out that there were no null values at all . We also used any() function to find out any null values for any variables bu there weren't any

### Then we found out whether there are any improper data types ,such as evapotranspiration having a data type as object. But we found none.



# Analysis technique:

## Through means of linear regression, we would like to investigate which of the attributes affect the observed variable the most. Specifically, we want to find attributes that best fit a linear regression line when plotted with the observed baseflow. Looking at the R^2 and p values will tell us if changes in any of these attributes lead to reliable and significant changes in our dependent variable, the baseflow.

## We will first examine these on the dataset as a whole, and then perform the analysis on individual segments and groups of segments.

### R-Square Values for every Segment-Id

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
pd.options.display.float_format = '{:.3f}'.format
lm=LinearRegression()
tableforRSquared=pd.DataFrame()
for each in df.Segment_id.unique():
    dfsegmentID = df.groupby('Segment_id').get_group(each)
    X=dfsegmentID.iloc[:,:-1]
    Y=dfsegmentID.iloc[:,-1]# Observed Series
    lm = smf.ols(formula = 'Observed ~ Date + Segment_id + x + y + Evapotranspiration + \
                            Precipitation + Irrigation_pumping',data=dfsegmentID).fit()
    tempTable = pd.concat([pd.Series(each),pd.Series(lm.rsquared)],axis=1)
    tableforRSquared=pd.concat([tableforRSquared,tempTable],axis=0)
print('The Statistical Summary for R-Squared-Values:')
tableforRSquared.columns=['Segment_Id','R-Square-values']
tableforRSquared.set_index('Segment_Id',inplace=True)
#tableforRSquared.sort_values(by='R-Square-values',ascending=False,inplace=True)
display(tableforRSquared.head(10))
```

## Now let's find out the  pvalues for each of the segment-Ids

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
pd.options.display.float_format = '{:.3f}'.format
lm=LinearRegression()
tableforPValues=pd.DataFrame()
for each in df.Segment_id.unique():
    dfsegmentID = df.groupby('Segment_id').get_group(each)
    X=dfsegmentID.iloc[:,:-1]
    Y=dfsegmentID.iloc[:,-1]# Observed Series
    temp=pd.DataFrame()
    lm = smf.ols(formula = 'Observed ~ Date + Segment_id + x + y + Evapotranspiration + \
                            Precipitation + Irrigation_pumping',data=dfsegmentID).fit()
    temp = pd.concat([temp,pd.Series(each)],axis=1)
    for elem in lm.pvalues:
        temp= pd.concat([temp,pd.Series(elem)],axis=1)
    tableforPValues = pd.concat([tableforPValues,temp],axis=0)
tableforPValues.columns=['Segment_Id','P-value:Intercept','P-value:Date','P-value:Segment_id','P-value:x','P-value:y',
                         'P-value:EvapoTransportation','P-value:Precipitation','P-value:Irrigation-Pumping']
tableforPValues.set_index('Segment_Id',inplace=True)
tableforPValues.head(10)
```

```python
import math

def distance_function(table, id_1, id_2):
    if id_1 == id_2: 
        return 0
    x1 = table[table.Segment_id == id_1].iloc[0].x
    y1 = table[table.Segment_id == id_1].iloc[0].y
    x2 = table[table.Segment_id == id_2].iloc[0].x
    y2 = table[table.Segment_id == id_2].iloc[0].y
    return int(math.sqrt(((x2 - x1) ** 2) + ((y2 - y1) ** 2)))
```

```python
segment_ids = sorted(df.Segment_id.unique().tolist())

distance_columns = ['id'] + ['distance_from_' + str(id) for id in segment_ids]
    
distance_rows = [[id_1] + [distance_function(df, id_1, id_2) for id_2 in segment_ids] for id_1 in segment_ids]
```

```python
Distancetable = pd.DataFrame(distance_rows, columns=distance_columns)
```

```python
# Functions to make filtering and displaying the tables easier
def filter_function(table, p, coef, r):
    return table[(table['p_Value'] < p) & (abs(table['Coefficient_Value']) > coef) & (table['R_Squared_Value'] > r)]

def filter_tables(tables, p, coef, r):
    for table_name, table in tables.items():
        display(table_name + ":")
        display(filter_function(table, p, coef, r))
```
